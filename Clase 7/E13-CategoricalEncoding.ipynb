# Exercise 13

This particular Automobile Data Set includes a good mix of categorical values as well as continuous values and serves as a useful example that is relatively easy to understand. Since domain understanding is an important aspect when deciding how to encode various categorical values - this data set makes a good case study.

Read the data into Pandas

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
import numpy as np
import category_encoders as ce
```

```python
# Define the headers since the data does not have any
headers = ["symboling", "normalized_losses", "make", "fuel_type", "aspiration",
           "num_doors", "body_style", "drive_wheels", "engine_location",
           "wheel_base", "length", "width", "height", "curb_weight",
           "engine_type", "num_cylinders", "engine_size", "fuel_system",
           "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm",
           "city_mpg", "highway_mpg", "price"]

# Read in the CSV file and convert "?" to NaN
df = pd.read_csv("http://mlr.cs.umass.edu/ml/machine-learning-databases/autos/imports-85.data",
                  header=None, names=headers, na_values="?" )
df.head()
```

# Exercise 13.1

Does the database contain missing values? If so, replace them using one of the methods explained in class

```python
df.dropna(subset=['price'],inplace = True)
df.shape
```

```python
# check for missing values
df.isnull().sum()
```

```python
df.normalized_losses.describe()
```

```python
# fill missing values for Age with the median age
df.normalized_losses.fillna(df.normalized_losses.median(), inplace=True)
df.isnull().sum()
df.dropna().shape
```

```python
dict_cilindros = {
    'four':4,
    'six':6,
    'five':5,
    'three':3,
    'twelve':12,
    'two':2,
    'eight':8
}
df.replace({'num_cylinders':dict_cilindros}, inplace=True)
```

```python
df['num_cylinders'].unique()
```

```python
dict_puertas = {
    'two':2,
    'four':4,
}
df.replace({'num_doors':dict_puertas}, inplace=True)
df.num_doors.fillna(4, inplace=True)
```

```python
df['num_doors'].unique()
```

```python
df.loc[df.horsepower.isnull()]
```

```python
df.dropna(inplace=True)
df.shape
```

```python
# check for missing values
df.isnull().sum()
```

# Exercise 13.2

Split the data into training and testing sets

Train a Random Forest Regressor to predict the price of a car using the nominal features

```python
df2 = df.select_dtypes(include=['float64','int64']).copy()
df2.head()
```

```python
Y = df2['price']
X = df2.drop(['price'], axis=1)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=1)
```

```python
rf = RandomForestRegressor(n_estimators=100,max_depth=None,max_features=None, random_state=1)
rf.fit(X_train,Y_train)
```

```python
y_pred = rf.predict(X_test)
print('MAE:', metrics.mean_absolute_error(Y_test, y_pred))
print('MSE:', metrics.mean_squared_error(Y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, y_pred)))
```
